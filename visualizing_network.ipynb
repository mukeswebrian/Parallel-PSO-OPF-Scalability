> ### Algorithm to covert result network into DOT graph format

```python
transformers = network_data[1]
lines = network_data[2]
loads = network_data[3]
gens = network_data[5]
res_lines = network_data[9]
res_trafo = network_data[9]
```

```python
def load_transformer_edges(dot_string, transformers):
    for i in transformers.index:
        n1 = transformers.hv_bus.loc[i]
        n2 = transformers.lv_bus.loc[i]
        
        dot_string += f'\n{n1} -- {n2} [color=red];'
        
    return dot_string

def load_line_edges(dot_string, lines):
    for i in lines.index:
        n1 = lines.from_bus.loc[i]
        n2 = lines.to_bus.loc[i]      
        
        dot_string += f'\n{n1} -- {n2} [color=blue];'
        
    return dot_string

def add_load_bus_attributes(dot_string, loads):
    complete = set()
    
    for i in loads.bus:
        if i not in complete:
            dot_string += f'\n{i}[shape=box]'
            
    return dot_string

def add_gen_bus_attributes(dot_string, gens):
    complete = set()
    
    for i in gens.bus:
        
        if i not in complete:
            dot_string += f'\n{i}[color=green]'
            
    return dot_string
    
```

```python
dot_string = 'graph mynet {'
dot_string = load_transformer_edges(dot_string, transformers)
dot_string = load_line_edges(dot_string, lines)
dot_string = add_load_bus_attributes(dot_string, loads)
dot_string = add_gen_bus_attributes(dot_string, gens)
dot_string += '\n}'
print(dot_string)
```
