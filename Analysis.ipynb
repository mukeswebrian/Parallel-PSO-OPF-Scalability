```python
import pandas as pd

table = pd.read_excel('run_plan.xlsx', sheet_name='Sheet1')
#table.min_runtime = table.min_runtime.apply(math.log)
table
```

```python
ax_title_fmt = {
    'size':14
}

tick_fmt = {
    'size':'large'
}

legend_fmt = {
    'fontsize':14
}

title_fmt = {
    'size':16
}

text_fmt = {
    'fontsize':9,
}
```

```python
import matplotlib.pyplot as plt
import math

fig, ax = plt.subplots(figsize=(10,8))


to_plot = table.groupby('particles')
to_plot.plot(kind='line', x='processes', y='min_runtime', ax=ax)
plt.grid()
plt.legend(['48 particles', '96 particles', '192 particles'], **legend_fmt)
plt.ylabel('duration (seconds)', fontdict=ax_title_fmt)
plt.xlabel('number of processes', fontdict=ax_title_fmt)
plt.title('Minimum Run Duration vs Number of Processes', **title_fmt)
plt.yticks(**tick_fmt)
plt.xticks(**tick_fmt)
```
